# 1.Read the dataset to python envirorment.

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
#read the dataset Iris
filename =pd.read_excel('Datasets-20211106\iris.xls')
df = pd.DataFrame(filename)
df
```

# 2.Display the columns in the dataset.

```python
list(df.columns)
```

# 3.Calculate the mean of each column of the dataset.

```python
#find the mean of all columns except Classification 
df[['SL','SW','PL','PW']].mean()
```

```python
print('Mean of SL is: ',round(df['SL'].mean(),2))
print('Mean of SW is: ',round(df['SW'].mean(),2))
print('Mean of PL is: ',round(df['PL'].mean(),2))
print('Mean of PW is: ',round(df['PW'].mean(),2))
```

# 5.Perform meaningful visualizations using the dataset. Bring at least 3.

# *)Count plot

```python
df['Classification'].value_counts()
```

```python
plt.style.use('seaborn')
sns.countplot(df['Classification'])
plt.title('Count plot of Classification',fontsize=18)
plt.xlabel('Classification',fontsize=16)
plt.ylabel('Count',fontsize=16)
plt.tight_layout()
```

# *)Pair plot

```python
plt.style.use('seaborn')
sns.pairplot(df)
```

# *) Sepal length in different species of Iris

```python
#plot a scatter plot between sepal_lengths of different Iris species
plt.style.use('seaborn')
plt.figure(figsize=(8,8))
sns.stripplot(x=df['Classification'], y=df['SL'])
plt.title('Sepal length in different species of Iris',fontsize=16)
plt.xlabel('Classification',fontsize=14)
plt.ylabel('Sepal_length',fontsize=14)
plt.tight_layout()
```

# *)Sepal width in different species of Iris

```python
#plot a scatter plot between sepal_widths of different Iris species
plt.style.use('seaborn')
plt.figure(figsize=(8,8))
sns.stripplot(x=df['Classification'], y=df['SW'])
plt.title('Sepal width in different species of Iris',fontsize=16)
plt.xlabel('Classification',fontsize=14)
plt.ylabel('Sepal_width',fontsize=14)
plt.tight_layout()
```

# *)Petal length in different species of Iris

```python
#plot a scatter plot between petal_lengths of different Iris species
plt.style.use('seaborn')
plt.figure(figsize=(8,8))
sns.stripplot(x=df['Classification'], y=df['PL'])
plt.title('Petal length in different species of Iris',fontsize=16)
plt.xlabel('Classification',fontsize=14)
plt.ylabel('Petal_length',fontsize=14)
plt.tight_layout()
```

# *)Petal Width in different species of Iris

```python
#plot a scatter plot between petal_widths of different Iris species
plt.style.use('seaborn')

sns.stripplot(x=df['Classification'], y=df['PW'])
plt.title('Petal width in different species of Iris',fontsize=16)
plt.xlabel('Classification',fontsize=14)
plt.ylabel('Petal_width',fontsize=14)
plt.tight_layout()
```

# *)Relation between Petal length and Petal width of 3 Iris species

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,8))
sns.scatterplot(x='PL',y='PW',data=df,hue='Classification')
plt.title('Relationship between Petal length & Petal width of 3 Iris species',fontsize=14)
plt.xlabel('Petal length',fontsize=14)
plt.ylabel('Petal width',fontsize=14)
plt.tight_layout
```

# *)Relation between sepal length and sepal width of 3 Iris species

```python
plt.style.use('seaborn')
sns.scatterplot(x='SL',y='SW', data=df, hue='Classification')
plt.title('Relationship between sepal length & sepal width of 3 Iris species',fontsize=14)
plt.xlabel('Sepal length',fontsize=12)
plt.ylabel('Sepal width',fontsize=12)
plt.tight_layout()
```

# *)Heatmap on Correlation of features

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(),annot=True)
plt.title('Heatmap on Correlation of features',fontsize=16)
plt.tight_layout()
```
